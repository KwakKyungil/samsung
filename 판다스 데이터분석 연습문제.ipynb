## 3. 데이터프레임 합치기(merge)
* 문제 1: 주어진 두 데이터프레임(df1, df2)을 '학생ID'를 기준으로 데이터프레임을 합쳐보세요.

```python
import pandas as pd

data1 = {
    '학생ID': [1, 2, 3, 4, 5],
    '이름': ['철수', '영희', '민수', '수지', '지영'],
    '학년': [1, 2, 2, 3, 1]
}

data2 = {
    '학생ID': [1, 2, 3, 4, 5],
    '국어': [80, 90, 70, 85, 95],
    '영어': [90, 80, 70, 60, 85]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, on = '학생ID')
```
